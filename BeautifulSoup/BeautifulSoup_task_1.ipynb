#  BeautifulSoup:

```python
file = open("html_doc.txt")
html_doc = file.read() 
html_doc
```

## Import BeautifulSoup4:

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc,'html.parser')
```

## prettify()
#### The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each tag and each string:

```python
print(soup.prettify())
```

## title tag:

### .title:
#### It gives whole tag with values
### .title.name:
#### It gives only the name of html tag
### .title.name.string:
#### It gives only the string between the html tag
### .title.parent.name:
#### It gives  the parent tag of title tag

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.title.parent.name
```

## all links used in anchor tag:

```python
for link in soup.find_all('a'):
    print(link['href'])
```

## all image sources used in img tag:

```python
for src in soup.find_all('img'):
    print(src['src'])
```

## all span tag:

```python
spans = soup.find_all('span')
spans
```

## all tags:

```python
tags = soup.find_all(['div','span','a','li'])
tags
```
